```python
kmeans = KMeans(n_clusters = 3,random_state=42)
k_fit = kmeans.fit(X_pca)
```

```python
sets = k_fit.labels_
```

```python
# Let's visualize the data we reduced to 4 sets.

plt.scatter(X_pca.iloc[:,0], X_pca.iloc[:,1], c = sets, s = 100, cmap = "viridis")

centers = k_fit.cluster_centers_                                
plt.scatter(centers[:,0], centers[:,1], c = "black", s = 300, alpha = 0.5);
```

```python
kmeans_2 = KMeans(n_clusters = 2,random_state=42)
k_fit_2 = kmeans_2.fit(X_pca)
sets_2 = k_fit_2.labels_
centers_2 = kmeans_2.cluster_centers_
```

# Hierarchy


```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
wine_2=pd.read_csv('wine.csv')
wine_2.head()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(wine_2.iloc[:,1:])
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
hc
```

```python
hc = AgglomerativeClustering(n_clusters=2, affinity = 'euclidean', linkage = 'single')
hc
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```
